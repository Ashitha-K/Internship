```python
#Handling missing values of categorical columns
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='most_frequent')
data['Gender']=imputer.fit_transform(data[['Gender']]).ravel()
data['Payment Method']=imputer.fit_transform(data[['Payment Method']]).ravel()
data['Preferred Visit Time']=imputer.fit_transform(data[['Preferred Visit Time']]).ravel()
data['Membership Status']=imputer.fit_transform(data[['Membership Status']]).ravel()
```

```python
#handling missing values of numerical columns
imputer=SimpleImputer(strategy='mean')
data['Store Rating']=imputer.fit_transform(data[['Store Rating']]).ravel()
```

```python
#Checking outliers
# Calculate IQR for numerical columns
continous_columns=['Age','Store Rating','Total Spent','Discount (%)','Revenue']
Q1 = data[continous_columns].quantile(0.25)
Q3 = data[continous_columns].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = ((data[continous_columns] < lower_bound) | (data[continous_columns] > upper_bound)).sum()
print(outliers)
```

```python
#Checking for outliers
outlier_columns=['Store Rating','Total Spent','Revenue']
for i in outlier_columns:
  plt.figure(figsize=(5, 4))
  sns.boxplot(data[i])
  plt.show()
```

```python
#Clipping  the outliers
for i in continous_columns:
  data[i]=np.where(data[i]>upper_bound[i],upper_bound[i],data[i])
  data[i]=np.where(data[i]<lower_bound[i],lower_bound[i],data[i])
```

```python
#Plotting the outlier after clipping
for i in outlier_columns:
  plt.figure(figsize=(5, 4))
  sns.boxplot(data[i])
  plt.show()
```

```python
#Correlation separating the numerical columns
data_corr=data.select_dtypes(include=['int64','float64'])
data_corr.corr()
ax=plt.subplots(figsize=(10,10))
sns.heatmap(data_corr.corr(),annot=True)

```

Total Spent and Discount (%): These two variables show a strong negative correlation (-0.98). This suggests that as the discount percentage increases, the total amount spent decreases, likely because higher discounts reduce the overall cost.

Discount (%) and Revenue: These two variables have a strong negative correlation (-0.83), implying that as discount rates go up, revenue tends to decrease, which is logical as higher discounts reduce revenue per item.

Satisfaction Score and Loyalty Score: There is a strong positive correlation (0.94) between satisfaction and loyalty, suggesting that satisfied customers are more likely to have higher loyalty scores.

Revenue and Total Spent: These are highly positively correlated (0.88), meaning that higher spending directly contributes to revenue.

Revenue and Loyalty Score: There is a significant positive correlation (0.81) between revenue and loyalty score, indicating that loyal customers tend to spend more, contributing positively to revenue.

Items Purchased and Discount (%): These two variables have a strong negative correlation (-0.85), suggesting that higher discounts might encourage customers to buy more items.

Satisfaction Score and Revenue: The correlation is moderate (0.79), implying that higher customer satisfaction is associated with an increase in revenue.
